```python
def computador_escolhe_jogada(n, m):
    resposta = n % (m + 1)
    if resposta == 0:
        return m if m > n else n
    else:
        return resposta
```

```python
computador_escolhe_jogada(1, 3)
```

```python
def usuario_escolhe_jogada(n,m):
    resposta = int(input("Quantas peças você vai tirar? "))
    print("")
    while (resposta < 1 or resposta > m):
        print("Oops! Jogada inválida! Tente de novo.\n")
        resposta = int(input("Quantas peças você vai tirar? "))
    return resposta
```

```python
usuario_escolhe_jogada(10, 3)
```

```python
def partida():
    dic = {
        1: 'uma peça',
        2: 'duas peças',
        3: 'três peças',
        4: 'quatro peças',
        5: 'cinco peças',
        6: 'seis peças',
        7: 'sete peças',
        8: 'oito peças',
        9: 'nove peças'
        
    }
    print("Bem-vindo ao jogo do NIM! Escolha: ")

    print("1 - para jogar uma partida isolada ")
    op = int(input("2 - para jogar um campeonato "))

    if op == 2:
        print("\nVocê escolheu um campeonato!\n")
        vitorias_pc, vitorias_usu = 0, 0
        for partida in range(1, 4):
            print("**** Rodada " + str(partida) + " ****\n")
            n = int(input("Quantas peças? "))
            m = int(input("Limite de peças por jogada: "))
            restam = n
            iniciante = -1
            if (n % (m + 1) == 0):
                print("\nVocê começa!\n")
                usu = usuario_escolhe_jogada(restam,m)
                print("\nVocê tirou", dic[usu])
                restam -= usu 
                iniciante = 0
            else:
                print("\nComputador começa!\n")
                pc = computador_escolhe_jogada(restam,m)
                print("O computador tirou", dic[pc])
                restam -= pc
                iniciante = 1
            while restam > 0:
                if restam == 1:
                    print("Agora resta apenas uma peça no tabuleiro.\n")
                else: 
                    print("Agora restam", restam, "peças no tabuleiro.\n")
                if iniciante == 1:
                    usu = usuario_escolhe_jogada(restam,m)
                    print("\nVocê tirou", dic[usu])
                    restam -= usu 
                    iniciante = 0
                else:
                    pc = computador_escolhe_jogada(restam,m)
                    print("O computador tirou", dic[pc])
                    restam -= pc
                    iniciante = 1
            if (iniciante == 1):
                print("Fim do jogo! O computador ganhou!\n")
                vitorias_pc += 1
            else:
                print("Fim do jogo! Você ganhou!\n")
                vitorias_usu += 1
        print("**** Final do campeonato! ****\n")
        print("Placar: Você", vitorias_usu, "X", vitorias_pc, "Computador")
            
    else:
        print("Você escolheu uma partida!")
        n = int(input("Quantas peças? "))
        m = int(input("Limite de peças por jogada: "))
        restam = n
        iniciante = -1
        if (n % (m + 1) == 0):
            print("\nVocê começa!\n")
            usu = usuario_escolhe_jogada(restam,m)
            print("\nVocê tirou", dic[usu])
            restam -= usu 
            iniciante = 0
        else:
            print("\nComputador começa!\n")
            pc = computador_escolhe_jogada(restam,m)
            print("O computador tirou", dic[pc])
            restam -= pc
            iniciante = 1
        while restam > 0:
            if restam == 1:
                print("Agora resta apenas uma peça no tabuleiro.\n")
            else: 
                print("Agora restam", restam, "peças no tabuleiro.\n")
            if iniciante == 1:
                usu = usuario_escolhe_jogada(restam,m)
                print("\nVocê tirou", dic[usu])
                restam -= usu 
                iniciante = 0
            else:
                pc = computador_escolhe_jogada(restam,m)
                print("O computador tirou", dic[pc])
                restam -= pc
                iniciante = 1
        if (iniciante == 1):
            print("Fim do jogo! O computador ganhou!\n")
        else:
            print("Fim do jogo! Você ganhou!\n")
```

```python
partida()
```
